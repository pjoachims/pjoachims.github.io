```python
import yaml
import pyperclip
    
def work_to_html(work_dict: dict, work_abbrev: list = None) -> str:
    def _work_to_html(data: dict, s="  "):
        out = ""
        # Full row
        out += "<div class='cv_elem_flex'>\n"
        
        # Left side
        out += s + "<div class='date_loc'>\n"
        out += s*2 + f"<span class='work_date'> {data['date']['start']} - {data['date']['end']} </span>\n"
        out += s*2 + "<br>\n"
        out += s*2 + f"<span class='work_loc'> {data['company']['city']}, {data['company']['country']}</span>\n"
        out += s + "</div>\n"
            
        # Right side
        out += s + "<div class='work_right'>\n"
        # Right side -> summary line
        out += s*2 + f"<p class='elem_summary'>\n"
        out += s*2 + f"<span class='pos'> {data['position']} </span>\n"
        out += s*2 + f"<span class='comp'> @ {data['company']['name']} - {data['company']['department']} </span>\n"
        out += s*2 + "</p>\n"
        # Right side -> Bullets
        out += s*2 + "<div class='work_details'>\n"
        out += s*3 + "<ul>\n"
        for b in data["bullets"]:
            out += s*4 + "<li>\n"
            out += s*5 + b + "\n"
            out += s*4 + "</li>\n"
        out += s*3 + "</ul>\n"
        # Close bullets
        out += s*2 + "</div>\n"
        # Close work_right
        out += s + "</div>\n"
        # Close cv_elem_flex
        out += "</div>\n"
        return out
    
    out = ""
    if work_abbrev is None:
        for v in work_dict.values():
            out += _work_to_html(v)
    else:
        for v in work_abbrev:
            out += _work_to_html(work_dict[v]) 
    pyperclip.copy(out)
    return out

def uni_to_html(edu_dict: dict) -> str:
    def _uni_to_html(data: dict, s="  "):
        out = ""
        
        # Full row
        out += "<div class='cv_elem_flex'>\n"
        
        # Left side -> date, location
        out += s + "<div class='date_loc'>\n"
        out += s*2 + f"<span class='edu_date'>{data['date']['start']} - {data['date']['end']}</span>\n"
        out += s*2 + "<br>\n"
        out += s*2 + f"<span class='edu_loc'>{data['city']}, {data['country']}</span>\n"
        out += s + "</div>\n"
        
        degree = data["degree"]
        # Right side
        out += s + "<div class='edu_right'>\n"
        # Right side -> header
        out += s*2 + f"<p class='elem_summary'>\n"
        out += s*3 + f"<span class='degree_deg'>{degree['deg']}</span>\n"
        out += s*3 + f"<span class='degree_name'>{degree['name']}</span>\n"
        out += s*3 + f"<span class='comp'> @ {data['name']}</span>\n"
        out += s*2 + "</p>\n"
        # Right side -> edu_details
        out += s*2 + "<div class='edu_details'>\n"
        # Right side -> education details: grade
        out += s*3 + f"<span class='edu_det_cat'>Ø Grade: </span>\n"
        out += s*3 + f"<span class='edu_det_grade'>{data['grade']}</span>\n"
        out += s*3 + "<br>\n"
        # Right side -> education details: thesis
        t = data['thesis']
        out += s*3 + f"<span class='edu_det_cat'>Title of Thesis: </span>\n"
        out += s*3 + f"<i class='edu_thesis_title'>{t['name']}</i>\n"
        links = t["links"]
        if links != "None":
            for k, v in links.items():
                out += s*2 + f"<a href={v}>[{k}]</a>\n"
        # Close edu_details
        out += s*2 + "</div>\n"
        # Close edu_right
        out += s + "</div>\n"
        
        # Close cv_elem_flex
        out += "</div>\n"
        
        return out
    
    out = ""
    for v in edu_dict.values():
        out += _uni_to_html(v) 
    pyperclip.copy(out)
    return out
    
    
with open("cv.yaml", "r") as stream:
    data = yaml.safe_load(stream)
# work_to_html(data["work"], ['sra_huor', 'intern_pwc', 'sa_ecapital'])
uni_to_html(data["education"]["uni"])
```
